```python
import re  
import pandas as pd
```

```python
f=open('Whatsapp Chat with CA$H.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data[:850])
print(type(data))
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},?\s\d{1,2}:\d{2}\s[APMapm]{2}\s-\s'
#pattern='\d{1,2}\/\d{1,2}\/\d{2},\s\d{1,2}:\d{2}\s(?:AM|PM)'     
```

```python
messages = re.split(pattern,data)[1:]
messages[:100]
```

```python
dates = re.findall(pattern,data)
print(dates)
```

```python
df = pd.DataFrame({'user_message':messages , 'message_date': dates})
#conv message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
print(df.columns)
```

```python
df.shape
```

```python
#seperate users and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message, maxsplit=1)  # Updated regular expression and added maxsplit
    if len(entry) > 1:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group-notification')
        messages.append(entry[0])

# Update column names
df['user'] = users
df['message'] = messages

# Drop the correct column
df.drop(columns=['user_message'], inplace=True)

# Display the first few rows of the DataFrame
df.head()
```

```python
df['year'] =df['date'].dt.year
df['month'] =df['date'].dt.month_name()
df['day'] =df['date'].dt.day
df['hour'] =df['date'].dt.hour
df['minute'] =df['date'].dt.minute



df.head()
```
